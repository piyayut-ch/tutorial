```r
thai_date <- function(x, fmt = NULL) {
  y <- (year(x) + 543) %>% as.character()
  m <- month(x) %>% as.character()
  d <- day(x) %>% as.character()
  
  dt <- paste(d, m, y, sep = "-")
  
  return(dt)
}
```

```r
last_month <- Sys.Date() - 0:29
df <- data.frame(
  date = last_month,
  price = runif(30)
)
base <- ggplot(df, aes(date, price)) +
  geom_line()

# The date scale will attempt to pick sensible defaults for
# major and minor tick marks. Override with date_breaks, date_labels
# date_minor_breaks arguments.
base + scale_x_date(labels = thai_date)
```
